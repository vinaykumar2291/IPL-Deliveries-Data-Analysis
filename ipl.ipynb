```python
import pandas as pd
df = pd.read_csv("deliveries.csv")
print(df.head())
```

```python
 #Dataset information
print("\nDataset Info")
print(df.info())
```

```python
# Total runs scored
total_runs = df["runs_of_bat"].sum()
print("\nTotal Runs in Dataset:", total_runs)
```

```python
# Total runs scored
total_runs = df["runs_of_bat"].sum()
print("\nTotal Runs in Dataset:", total_runs)
```

```python
# Top 10 batsmen by runs
top_batsmen = df.groupby("striker")["runs_of_bat"].sum().sort_values(ascending=False).head(10)
```

```python
print("\nTop 10 Batsmen")
print(top_batsmen)
```

```python
# Top 10 bowlers by wickets
wickets = df[df["player_dismissed"].notna()]
top_bowlers = wickets.groupby("bowler")["player_dismissed"].count().sort_values(ascending=False).head(10)
```

```python
print("\nTop 10 Bowlers")
print(top_bowlers)
```

```python
# Matches per venue
venues = df["venue"].value_counts().head(10)
```

```python
print("\nTop Venues")
print(venues)

```
